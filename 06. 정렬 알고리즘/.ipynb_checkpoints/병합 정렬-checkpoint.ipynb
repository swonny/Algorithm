````python
# ì •ë ¬ì„ ë§ˆì¹œ ë‘ ë°°ì—´ mergeí•˜ê¸° - ë‚´ê°€ ì§  ì½”ë“œ(ë‹¤ë¦„) : ì±…ì€ íŒŒì´ì¬ì˜ ìŠ¬ë¼ì´ìŠ¤ ê¸°ëŠ¥ ë“±ì„ ì´ìš©í•˜ì§€ ì•Šê³  í’€ì—ˆë‹¤. ë‹¤ì‹œí•´ë³´ì!

from typing import Sequence, MutableSequence

def merge_sort_list(a: Sequence, b: Sequence, c):
    pa, pb, pc = 0, 0, 0
    na, nb = len(a), len(b)
    nc = na + nb
    
    for i in range(nc):
        if pa == na:
            c[pc:] = b[pb:]
            break
        if pb == nb:
            c[pc:] = a[pa:]
            break
        if(a[pa] < b[pb]):
            c[pc] = a[pa]
            pa += 1
        else:
            c[pc] = b[pb]
            pb += 1
        pc += 1
            
a = [2, 4, 6, 8, 11, 13]
b = [1, 2, 3, 4, 9, 16, 21]
c = [None] * (len(a) + len(b))
merge_sort_list(a, b, c)
print(c)
````
