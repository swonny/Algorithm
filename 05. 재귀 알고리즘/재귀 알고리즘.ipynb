````python
from collections import deque

class Stack:
    def __init__(self):
        self.top = 0
        self.stk = deque([])
    
    def is_empty(self):
        if self.top == 0:
            return True
    
    def push(self, value):
        self.stk.append(value)
        self.top += 1
    
    def pop(self):
        if not self.is_empty():
            self.top -= 1
            return self.stk.pop()
````

````python
# ì¬ê·€í˜¸ì¶œ ìŠ¤íƒì„ ì´ìš©í•˜ì—¬ ë¹„ì¬ê·€ì  í•¨ìˆ˜ë¡œ êµ¬í˜„í•˜ê¸° - ì§ì ‘ ì§  ì½”ë“œ

# ë°”ê¹¥ìª½ whileë¬¸ì˜ ì¢…ë£Œì¡°ê±´ì„ êµ¬í•˜ëŠ” ê²Œ ë¶€ì¡±í–ˆìŒ. & continueë¥¼ ì´ìš©í•˜ë©´ whileë¬¸ì„ ë‘ ë²ˆ ì“°ì§€ ì•Šì•„ë„ ë¨. ifì— í•´ë‹¹í•˜ëŠ” ê²ƒì€ ì¬ê·€ ì• ë¶€ë¶„ì´ë‹ˆê¹Œ.

stk = Stack()

def recur(n):
    cnt = 0
    while cnt < 100:
        cnt += 1
        if n > 0:
            stk.push(n)
            n = n -1
        else:
            while not stk.is_empty():
                n = stk.pop()
                print(n)
                n = n - 2
                break
        
recur(5)        
````
