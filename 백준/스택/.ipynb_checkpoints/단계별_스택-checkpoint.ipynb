```python
class Stack:
    def __init__(self):
        self.stk = []
        self.ptr = 0
    
    def push(self, value):
        self.stk.append(value)
        self.ptr += 1
        
    def empty(self):
        if self.ptr <= 0:
            return 1
        return 0
    
    def pop(self):
        if self.empty() == 1:
            return -1
        self.ptr -= 1
        return self.stk.pop()
    
    def size(self):
        return self.ptr
    
    def top(self):
        if self.empty() == 1:
            return -1
        return self.stk[self.ptr - 1]   

n = int(input())
stk = Stack()
for _ in range(n):
    tmp = input()
    try:
        command, num = tmp.split()
    except ValueError:
        command = tmp  
    
    if command == 'push':
        stk.push(num)
    
    elif command == 'pop':
        print(stk.pop())
            
    elif command == 'size':
        print(stk.size())
    
    elif command == 'empty':
        print(stk.empty())
        
    elif command == 'top':
        print(stk.top())
```

```python
from sys import stdin

n = int(input())
stk = []

for _ in range(n):
    stk = []
    p = list(map(str, stdin.readline().rstrip()))
    for pa in p:
        if pa == '(':
            stk.append('(')
        else:
            try:
                stk.pop()
            except IndexError:
                print('NO')
                break
    else: 
        print('YES') if len(stk) == 0 else print('NO')
```

### 4949 Í∑†ÌòïÏû°ÌûåÏÑ∏ÏÉÅ

[Î©îÎ™®]
- c in 'string' ÏÇ¨Ïö© Í∞ÄÎä•

```python
from sys import stdin

def isBal(string):
    stk = []
    for e in string:
        if e == '(' or e == '[':
            stk.append(e)
        elif e == ')':
            if len(stk) == 0 or stk.pop() == '[':
                return 'no'
        elif e == ']':
            if len(stk) == 0 or stk.pop() == '(':
                return 'no'
    if len(stk): return 'no'
    return 'yes'

while True:
    string = str(stdin.readline().rstrip())
    if string == '.':
        break
    else:
        print(isBal(string))
```

```python
# 1Îì± ÏΩîÎìú
from sys import stdin, stdout

def isvalid(s):
    stack = []
    for c in s:
        if c in '([':
            stack.append(c)
        elif c == ')':
            if not stack or stack.pop() != '(':
                return False
        elif c == ']':
            if not stack or stack.pop() != '[':
                return False
    return not stack

strings = stdin.readlines()
strings.pop()
for string in strings:
    stdout.write("yes\n" if isvalid(string) else "no\n")
```

```python
from sys import stdin

n = int(input())
stk = []
ptr = 0 # ÏàòÏó¥ ptr
top = 0 # Ïä§ÌÉù Îç∞Ïù¥ÌÑ∞ Í∞úÏàò
result = []

numarr = list([input() for _ in range(n)])
for num in range(1, n+1):
    stk.append(num) # push
    top += 1
    result.append('+')

    while top > 0 and stk[top - 1] == int(numarr[ptr]):
        stk.pop()
        top -= 1
        result.append('-')
        ptr += 1

if(top > 0):
    stk.reverse()
    stkStr = ''.join(map(str, stk))
    numarrStr = ''.join(map(str, numarr[ptr:]))
    if stkStr == numarrStr:
        result.append(stkStr.split())
    else:
        print('NO')
        result= []

for n in result:
    print(n)
```

```python
# ÎãµÏùÄ ÎßûÏßÄÎßå ÏãúÍ∞Ñ Ï¥àÍ≥º
from sys import stdin

n = int(input())
arr = list(map(int, input().split()))
res = []
for i in range(n):
    tmp = [j for j in range(i+1, n) if arr[j] > arr[i]]
    res.append(-1 if not tmp else arr[min(tmp)]) 

print(' '.join(map(str, res)))
```
