```python
class Stack:
    def __init__(self):
        self.stk = []
        self.ptr = 0
    
    def push(self, value):
        self.stk.append(value)
        self.ptr += 1
        
    def empty(self):
        if self.ptr <= 0:
            return 1
        return 0
    
    def pop(self):
        if self.empty() == 1:
            return -1
        self.ptr -= 1
        return self.stk.pop()
    
    def size(self):
        return self.ptr
    
    def top(self):
        if self.empty() == 1:
            return -1
        return self.stk[self.ptr - 1]   

n = int(input())
stk = Stack()
for _ in range(n):
    tmp = input()
    try:
        command, num = tmp.split()
    except ValueError:
        command = tmp  
    
    if command == 'push':
        stk.push(num)
    
    elif command == 'pop':
        print(stk.pop())
            
    elif command == 'size':
        print(stk.size())
    
    elif command == 'empty':
        print(stk.empty())
        
    elif command == 'top':
        print(stk.top())
```

### 4949 ê· í˜•ì¡íŒì„¸ìƒ

```python
from sys import stdin

def isBal(string):
    stk = []
    for e in string:
        if e == '(' or e == '[':
            stk.append(e)
        elif e == ')':
            if len(stk) == 0 or stk.pop() == '[':
                return 'no'
        elif e == ']':
            if len(stk) == 0 or stk.pop() == '(':
                return 'no'
    if len(stk): return 'no'
    return 'yes'

while True:
    string = str(stdin.readline().rstrip())
    if string == '.':
        break
    else:
        print(isBal(string))
```

```python
# 1ë“± ì½”ë“œ
from sys import stdin, stdout

def isvalid(s):
    stack = []
    for c in s:
        if c in '([':
            stack.append(c)
        elif c == ')':
            if not stack or stack.pop() != '(':
                return False
        elif c == ']':
            if not stack or stack.pop() != '[':
                return False
    return not stack

strings = stdin.readlines()
strings.pop()
for string in strings:
    stdout.write("yes\n" if isvalid(string) else "no\n")
```

```python
# ê°œì„  ì½”ë“œ

from sys import stdin

n = int(stdin.readline())
arr = list(map(int, stdin.readline().split()))

stk = [] # í¬ê¸° ë¹„êµ ìŠ¤íƒ
res = [] # ê²°ê³¼ ì €ì¥ ìŠ¤íƒ : ì¶œë ¥ ì‹œ ë’¤ì§‘ì–´ì•¼ í•¨

idx = -1
while -idx <= len(arr):
    if not stk:
        res.append(-1)
    else:
        if stk[-1] > arr[idx]:
            res.append(stk[-1])
        else:
            stk.pop()
            continue
    stk.append(arr[idx])
    idx -= 1

res.reverse()        
print(' '.join(map(str, res)))
```

```python
# 1ë“± ì½”ë“œ

import sys
def sol():
    n,*arr = map(int,sys.stdin.read().split())
    ans = ['-1'] * n
    stack = []
    for i in range(n):
        while stack and arr[stack[-1]] < arr[i]:
            ans[stack.pop()] = str(arr[i])
        stack.append(i)
    print(' '.join(ans))
sol()
```
